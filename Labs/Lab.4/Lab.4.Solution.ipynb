1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, max_value, increment):
        if max_value <= 0:
            raise ValueError("Max value must be greater than 0")
        self.max_value = max_value
        self.value = 0
        self.increment = increment

    def incrementation(self):
        if self.value + self.increment <= self.max_value:
            self.value += self.increment
        else:
            self.value = self.max_value
            print("Max value reached")
            return False 
        return True

    def reset(self):
        self.value = 0

    def get_value(self):
        return self.value

    def is_at_max(self):
        return self.value == self.max_value

counter = Counter(600, 10)
for i in range(500):
    if not counter.incrementation():
        break
    print(counter.get_value())
    
print(counter.is_at_max())
counter.reset()
print(counter.get_value())
        
```

```python
from paint import Canvas, rectangle, circle, triangle
import math

class RasterDrawing:
    def __init__(self, canvas):
        self.canvas = canvas
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def remove_shape(self, shape):
        if shape in self.shapes:
            self.shapes.remove(shape)
        else:
            print("Shape not found")

    def paint(self):
        self.canvas.clear_canvas() 
        for shape in self.shapes:
            if isinstance(shape, rectangle):
                self.draw_rectangle(shape)
            elif isinstance(shape, circle):
                self.draw_circle(shape)
            elif isinstance(shape, triangle):
                self.draw_triangle(shape)
                
    def modify(self, shape, new_shape):
        index = self.shapes.index(shape)
        self.shapes[index] = new_shape
        
    def draw_rectangle(self, rect):
        for i in range(rect._rectangle__length):
            for j in range(rect._rectangle__width):
                self.canvas.set_pixel(rect._rectangle__y + j, rect._rectangle__x + i, '*')

    def draw_circle(self, circ):
        x_center, y_center = circ._circle__x, circ._circle__y
        radius = circ._circle__radius
        for x in range(x_center - radius, x_center + radius + 1):
            for y in range(y_center - radius, y_center + radius + 1):
                if (x - x_center) ** 2 + (y - y_center) ** 2 <= radius ** 2:
                    if 0 <= x < self.canvas.width and 0 <= y < self.canvas.height:
                        self.canvas.set_pixel(y, x, ')')

    def draw_triangle(self, tri):
        x, y = tri._triangle__x, tri._triangle__y
        a = tri._triangle__side1
        height = tri._triangle__side2 

        points = [(int(x), int(y)), (int(x + a), int(y)), (int(x + a // 2), int(y + height))]

        for i in range(3):
            start = points[i]
            end = points[(i + 1) % 3]
            self.canvas.line(start[0], start[1], end[0], end[1], char='*')
        
        for row in range(int(y), int(y + height)):
            start_x = int(x + (row - y) * (a / height))
            end_x = int(x + a - (row - y) * (a / height))
            for col in range(start_x, end_x + 1):
                self.canvas.set_pixel(row, col, '*')

    def save(self, filename):
        with open(filename, "w") as f:
            f.write(f"canvas_width = {self.canvas.width}\n")
            f.write(f"canvas_height = {self.canvas.height}\n")
            for shape in self.shapes:
                f.write(f"{repr(shape)}\n")

    def loader(filename):
        with open(filename, "r") as f:
            lines = f.readlines()

            # Extract canvas dimensions
            canvas_width = int(lines[0].split('=')[1].strip())
            canvas_height = int(lines[1].split('=')[1].strip())

            # Create the canvas using the dimensions
            canvas = Canvas(canvas_width, canvas_height)

            # Extract shape constructor code
            shapes_code = [line.strip() for line in lines[2:]]

        # Use eval to recreate the shapes from their constructor code
        shapes = [eval(code) for code in shapes_code]

        # Create the RasterDrawing object and add the shapes
        drawing = RasterDrawing(canvas)
        for shape in shapes:
            drawing.add_shape(shape)

        return drawing


# Example Usage:

canvas = Canvas(30, 30)
drawing = RasterDrawing(canvas)

# Add shapes to the drawing
rect = rectangle(7, 3, 5, 3)
circ = circle(10, 10, 21)
tri = triangle(10, 10, 6, 6, 6)

drawing.add_shape(rect)
drawing.add_shape(circ)
drawing.add_shape(tri)

# Save the drawing to a file
drawing.save("drawing.txt")

# Load the drawing from the file
loaded_paint = RasterDrawing.loader('drawing.txt')

# Display the loaded painting
print("Loaded painting:")
loaded_paint.paint()
loaded_paint.canvas.display()
```
